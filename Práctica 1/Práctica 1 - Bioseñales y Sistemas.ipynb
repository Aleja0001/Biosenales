1) Escribir una función que reciba un diccionario con las notas de los alumnos de un curso y devuelva 
una serie con la nota mínima, la máxima, media y la desviación típica.


```python
#Inicialmente creamos el diccionario con las notas de los estudiantes: 
notas = {
    "Sofía": 4.5,
    "Juan": 3.8,
    "Valeria": 4.6,
    "Pedro": 2.7,
    "Ana": 4.5,
    "Simón": 3.8,
    "María": 4.9,
    "Samuel": 5.0,
    "Mariana": 4.3,
    "Omar": 3.7
}
#Para esta función utilizaremos pandas, por tanto, realizamos la importación.

import pandas as pd

#Posteriormente creamos la función, esta recibirá como argumento el diccionario que se analizará:
def analisis_notas(diccionario):
    
    # Convertir el diccionario en una Serie de pandas para analizar de manera más ágil
    # y efectiva las notas con las funciones de esta librería.
    notas = pd.Series(diccionario)
    
    # Calculamos las estadísticas con las funciones respectivas en cada caso,
    # además damos un nombre a cada resultado para mostrarlos después de forma organizada.
    minimo = notas.min()
    maximo = notas.max()
    media = notas.mean()
    desviacion = notas.std()
    
    # Finalmente retornamos y mostramos los resultados organizadamente:
    return print(f"Nota mínima:{minimo}\n Nota máxima: {maximo}\n Media: {np.round(media,2)}\n Desviación estándar: {np.round(desviacion,2)}")

analisis_notas(notas)
```

2. Cálculo del Índice de Masa Corporal (IMC) con Pandas.


```python
# Cargar los datos desde el archivo "pacientes_VitalDB.csv".
# Usamos una ruta relativa esperando que el archivo este en la misma carpeta que el presente proyecto.
datos = pd.read_csv("pacientes_VitalDB.csv")
```

```python
# Mostrar las primeras y últimas filas del DataFrame para conocer la estructura de los datos.
primerasFilas = datos.head() #head: muestra las primeras 5 filas.
ultimasFilas = datos.tail() #tail: muestra las últimas 5 filas.

print(f'Primeras filas de los datos: \n {primerasFilas}')
print(f'\n\nÚltimas filas de los datos: \n {ultimasFilas}')
```

```python
# Hacer una descripción general del DataFrame.
# Utilizaremos dos funciones que nos dan características diferentes del DataFrame:
i = datos.info()
i2 = datos.describe()

print(i)
print(i2)
```

```python
# Calcular el IMC
# Después de haber explorado los datos, nos damos cuenta que la altura esta dada como los centímetros después de 1 metro,
# por tanto, primero se harán unos pequeños ajustes para dejar la altura en las unidades correctas para los cálculos (m).

datos["height"] = (datos["height"] +100)/100
datos["height"]
```

```python
# Igualmente inferimos que el peso está expresado en libras, por tanto haremos la conversión a kg:

datos["weight"] = datos["weight"]*0.453592
datos["weight"]
```

```python
# Aplicamos la fórmula del IMC:
IMC = datos["weight"] / (datos["height"] ** 2)

# Esto lo agregamos como una columna del DataFrame llamada 'BMI':
datos["BMI"] = IMC

datos["BMI"]
```

```python
# Finalmente vamos a categorizar según el IMC de cada paciente:

# Inicializamos una lista vacía donde posteriormente se agregará la categoría que corresponda según el rango donde se ubique el paciente.
categorias = []

#Creamos un ciclo for para incluir todas las condiciones y se evalúe en cada paciente del DataFrame
for valor in datos["BMI"]:
    
    # Se enlistan todas las condiciones con sus respectivas categorías
    if valor < 18.5:
        categorias.append("Low weight")
    
    elif valor < 25:
        categorias.append("Normal")
    
    elif valor < 30:
        categorias.append("Overweight")
    
    else:
        categorias.append("Obesity")

#Finalmente el ciclo se cerrará cuando el valor del IMC entre en alguna de las categorías y esta se asignará al respetivo paciente evaluado.
#E igualmente creamos una nueva columna donde se enlistarán las categorías:
datos["BMI_category"] = categorias
```
